# STEP 2 : Data Cleaning

```python
import pandas as pd
```

# - File path for saving the DFW Rainfall Analysis data frame as a CSV file


```python
path = "E:\Sushma Files\POWER BI TRAINING-Nov 23\Alex the analyst DA Bootcamp\WEB SCRAPING PROJECT\DFW_Rainfall_Analysis_1900_2023.csv"
```

# - Reading the DFW Rainfall Analysis data from the CSV file into a pandas DataFrame


```python
df = pd.read_csv(path)

# displaying the data frame
df
```

```python
# Displaying information about the DFW Rainfall Analysis DataFrame
df.info()
```

```python
# Dropping the 'Total' column from the DataFrame
df = df.drop(columns = 'Total')

# Displaying the updated/modified DataFrame
df
```

# - Replacing 'T' (trace amounts) with 0.00 in the DFW Rainfall Analysis DataFrame

```python
df = df.replace("T", 0.00)

# Displaying the updated/modified DataFrame
df
```

# - Extracting numeric columns and converting them to numeric data type in the DFW Rainfall Analysis DataFrame

```python
numeric_columns = df.columns[1:]
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')


# Displaying updated information about the DataFrame after numeric conversion
df.info()
```

# - Calculating total rainfall by summing values across monthly columns in the DFW Rainfall Analysis DataFrame


```python
df['Total_Rainfall'] = df.iloc[:,1:13].sum(axis=1)
```

```python
# Displaying the updated/modified DataFrame with "Total Rainfall" Column
df
```
