# 2. Time Based Analysis 

# Monthly Patterns  

# - Calculate the average rainfall for each month over all the years

```python
monthly_avg_rainfall = df.iloc[:,1:13].mean()
monthly_avg_rainfall = monthly_avg_rainfall.round(2)

# Displaying monthly average rainfall data 
monthly_avg_rainfall
```

# - Visualize the average monthly rainfall with a bar plot

```python
plt.figure(figsize=(10,6))
#monthly_avg_rainfall.plot(kind='bar',color='skyblue', edgecolor="black")
plt.bar(monthly_avg_rainfall.index,monthly_avg_rainfall,color='skyblue', edgecolor="black")
plt.title("Average Monthly Rainfall in DFW (1900-2023)")
plt.xlabel("Month")
plt.ylabel("Average Monthly Rainfall (inches)")
plt.show()
```

# - Visualize the average monthly rainfall with a line plot

```python
plt.figure(figsize=(10, 6))
plt.plot(monthly_avg_rainfall.index, monthly_avg_rainfall, marker='o', linestyle='-', color='b')
plt.title('Average Monthly Rainfall in DFW (1900-2023)')
plt.xlabel('Month')
plt.ylabel('Average Monthly Rainfall (inches)')
plt.grid(True)
plt.show()
```

# - Month with the highest average monthly rainfall (wettest month)

```python
wettest_month = monthly_avg_rainfall.idxmax()

print("The wettest month is : ",wettest_month)
```

# - Month with the lowest average monthly rainfall (driest month)

```python
Driest_month = monthly_avg_rainfall.idxmin()

print("The Driest month is : ",Driest_month)
```

# Yearly Patterns 

# - Visualize the Yearly Total Rainfall in a line plot

```python
plt.figure(figsize=(20,10))
plt.plot(df['Year'], df['Total_Rainfall'], marker='o', linestyle='-')
plt.title('Total Annual Rainfall Over the Years 1900-2023')
plt.xlabel('Year')
plt.ylabel('Total Annual Rainfall (inches)')
plt.grid(True)
plt.show
```

# - Identifying Driest Year 

```python
Driest_year = df.loc[df['Total_Rainfall'].idxmin()]

print("The Driest year between 1900 and 2023 is/was :",Driest_year['Year'])
```

# 3. Seasonal Analysis

# - Creating a new data frame with 4 seasons 
- winter: dec, jan, feb 
- spring: mar, apr, may 
- summer: jun, jul, aug 
- fall: sep, oct, nov

```python
# Displaying information about the DFW Rainfall Analysis DataFrame
df.info()
```

```python
# Creating a new data frame with seasons and sum of respected columns data 
Seasonal_Rainfall_Data = pd.DataFrame()
Seasonal_Rainfall_Data['Winter'] = df.iloc[:,[1,2,12]].sum(axis=1)
Seasonal_Rainfall_Data['Spring'] = df.iloc[:,[3,4,5]].sum(axis=1)
Seasonal_Rainfall_Data['Summer'] = df.iloc[:,[6,7,8]].sum(axis=1)
Seasonal_Rainfall_Data['Fall'] = df.iloc[:,[9,10,11]].sum(axis=1)
Seasonal_Rainfall_Data['Year'] = df.iloc[:,[0]]

# Displaying Seasonal Rainfall Dataframe
Seasonal_Rainfall_Data
```

# - Creating seasonal box plots 

```python
# Box plot visualization for seasonal rainfall data
plt.figure(figsize=(12, 6))
Seasonal_Rainfall_Data.boxplot(column=['Winter', 'Spring', 'Summer', 'Fall'], grid=False, showmeans=True)
plt.title('Box Plots of Seasonal Rainfall')
plt.ylabel('Rainfall (inches)')
plt.show()
```

# - Calculating and visualizing the Seasonal Average Rainfall 1900-2023

```python
# Calculating Seasonal Average Rainfall
Average_Rainfall_Season = Seasonal_Rainfall_Data.iloc[:,0:4].mean()

# Displaying Seasonal Average Rainfall
Average_Rainfall_Season
```

# - Bar plot for visualizing the average rainfall across seasons 

```python
plt.figure(figsize=(10,6))
plt.bar(Average_Rainfall_Season.index,Average_Rainfall_Season,color='skyblue', edgecolor="black")
plt.title('Average Rainfall Across Seasons over 1900-2023')
plt.xlabel('Season')
plt.ylabel('Average Rainfall (inches)')
plt.show()
```

# 4. Correlation Analysis
- This analysis aims to uncover relationships between monthly rainfall patterns

# - Calculating Correlation Coefficients

```python
correlation_matrix = df.iloc[:, 1:13].corr().round(2)
correlation_matrix
```

```python
#Extracting the correlation coefficient between Jan and Feb
correlation_jan_feb = correlation_matrix.loc["Jan", "Feb"]
print(f"Correlation between Jan and Feb: {correlation_jan_feb}")
```

# - Create a heatmap using Matplotlib

```python
plt.figure(figsize=(10, 8))

heatmap = plt.imshow(correlation_matrix, cmap="coolwarm", interpolation="nearest")

# Add colorbar
plt.colorbar(heatmap, fraction=0.046, pad=0.04)

# Set axis labels and ticks
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

# Set title
plt.title("Correlation Matrix of Monthly Rainfall in DFW 1900-2023")

plt.show()
```
